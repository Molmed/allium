# The annotation files align to our current GRCh38.103 reference genome version

```python
annot = pd.read_csv(path + 'annotateddata.csv', sep = ';')
genes = pd.read_csv(path + 'finalgenelength.csv', index_col = 'Unnamed: 0')
```

```python
annot
```

```python
annot2 = annot[annot.feature == 'gene']
annot2
```

## Keep only the genes used in our project (n = 19,774)

```python
annot3 = annot2[annot2.gene_id.isin(genes.index)][['gene_id', 'gene_name']]
annot3
```

```python
annot3.set_index('gene_id', inplace = True)
annot3 = annot3.reindex(genes.index)
annot3
```

```python
sum(annot3.index == genes.index)
```

# Check the missing genes

```python
np.setdiff1d(annot3.gene_name, df.Gene)
```

```python
len(np.setdiff1d(annot3.gene_name, df.Gene))
```

## Fill the missing genes with zeros 

```python
zeros = pd.DataFrame(columns = df.columns[1:], index = list(np.setdiff1d(annot3.gene_name, df.Gene)))
zeros.fillna(0, inplace = True)
```

```python
zeros
```

```python
df.set_index('Gene', inplace = True)
```

```python
newdf = pd.concat([df, zeros], axis = 0)
newdf.shape
```

```python
newdf
```

# First, create a matrix for the 19766 genes and then fix the other for the duplicates (>1 ENSEMBL ID for some genes)

```python
newdf2 = newdf[newdf.index.isin(annot3.gene_name)]
newdf2.shape
```

```python
annot4 = annot3.drop_duplicates('gene_name', keep = 'first')
annot4
```

### From gene name to ENSEMBL id

```python
namedict = pd.Series(annot4.index.values,index=annot4.gene_name).to_dict()
namedict
```

```python
len(namedict)
```

```python
largedf = newdf2.rename(index = namedict)
largedf.shape
```

# fix the extra duplicates (same GENE name but not ID)

```python
annot3[annot3.gene_name.duplicated(keep = False)]
```

```python
dups = annot3[annot3.gene_name.duplicated(keep = False)]
dups
```

```python
extra = dups[~dups.index.isin(annot4.index)]
extra
```

## The duplicates will receive the same value 

```python
dupsdf = newdf2[newdf2.index.isin(extra.gene_name)]
dupsdf
```

```python
dupdict = pd.Series(extra.index.values,index=extra.gene_name).to_dict()
dupdict
```

```python
dupsdf.rename(index = dupdict, inplace = True)
```

### Merge all and reindex according to the order of the 19774 genes

```python
finaldf = pd.concat([largedf, dupsdf])
finaldf
```

```python
finaldf = finaldf.reindex(genes.index)
finaldf
```
